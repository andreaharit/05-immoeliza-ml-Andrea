```python
# Replace NULLS in livingroom_surface
# Uses a average % of the living_area
mask = df[["kitchen_surface", "livingroom_surface"]].notnull().all(1)
has_kitchen_livingroom = df[mask]

#has_kitchen_livingroom.groupby("district").apply(lambda x:pd.Series({'%_k_area':(x.kitchen_surface.sum()/x.living_area.sum()),'mean_livr_area':(x.livingroom_surface.sum()/x.living_area.sum())}))
percent_k = has_kitchen_livingroom.kitchen_surface.sum()/has_kitchen_livingroom.living_area.sum() 
percent_l = has_kitchen_livingroom.livingroom_surface.sum()/has_kitchen_livingroom.living_area.sum() 
print (percent_k)
print(percent_l )
```

```python
mask = df["kitchen_surface"].isnull() |df["livingroom_surface"].isnull()
test_1 = df[mask]
test_1[["price", "kitchen_surface", "livingroom_surface"]]
```

```python

df['livingroom_surface'] = df.livingroom_surface.fillna(round(df.living_area*percent_l,0))
df['kitchen_surface'] = df.kitchen_surface.fillna(round(df.living_area*percent_k,0))
```
